# Opacity

```python
import numpy as np
import matplotlib.pyplot as plt

title = 'MaunaKea (Sept, Oct, Nov)'
pwv = '1'
T_unit = 'K'
PWV_unit = 'mm'
freq_unit = 'GHz'
theta_unit = 'deg'
tau_unit = 'np'

ZENITHS = ['0', '45', '60', '70']

for zenith in ZENITHS:
    data = np.load(f'./trop_h2o_Jacobian_{zenith}.out')
    plt.plot(data[:,0], data[:,1])

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'$\tau$ [np]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(ZENITHS, title="Zenith", loc=(1.01, 0.705))
plt.title(f'{title} [PWV={pwv}]')
```

# Raleigh-Jeans Temperature Brightness

```python
ZENITHS = ['0', '45', '60', '70']

for zenith in ZENITHS:
    data = np.load(f'./trop_h2o_Jacobian_{zenith}.out')
    plt.plot(data[:,0], data[:,2])

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'T$_b$ ' + f'[{T_unit}]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(ZENITHS, title="Zenith", loc=(1.01, 0.705))
plt.title(f'{title} [PWV={pwv}]')
```

# dTb / dZA

```python
for zenith in ZENITHS:
    data = np.load(f'./trop_h2o_Jacobian_{zenith}.out')
    plt.plot(data[:,0], data[:,3])

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'$\partial$T$_b$ / $\partial$$\theta$ ' + f'[{T_unit} / {theta_unit}]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(ZENITHS, title="Zenith", loc=(1.01, 0.705))
plt.title(f'{title} [PWV={pwv}]')
```

# dTb / dPWV

```python
correction_factor = 2109.86 * 10**-3

for zenith in ZENITHS:
    data = np.load(f'./trop_h2o_Jacobian_{zenith}.out')
    plt.plot(data[:,0], data[:,4] / correction_factor)

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'$\partial$T$_b$ / $\partial$PWV ' + f'[{T_unit} / {PWV_unit}]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(ZENITHS, title="Zenith", loc=(1.01, 0.71))
plt.title(f'{title} [PWV={pwv}]')

# dT / dPWV = dT / Nscale * (Nscale / mm_pwv)
```
