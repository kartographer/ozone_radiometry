# Opacity

```python
import numpy as np
import matplotlib.pyplot as plt

title = 'MaunaKea (Sept, Oct, Nov)'
pwv = '1'
T_unit = 'K'
PWV_unit = 'mm'
freq_unit = 'GHz'
theta_unit = 'deg'
tau_unit = 'np'

NSCALES = ['0.206', '0.513', '1.00', '2.07', '4.60']
common_name = 'trop_h2o_Jacobian'

for nscale in NSCALES:
    data = np.load(f'{common_name}_{nscale}.out')
    plt.plot(data[:,0], data[:,1])

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'$\tau$ ' + f'[{tau_unit}]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(NSCALES, title='NSCALE', loc=(1.01, 0.65))
plt.title(f'{title} [PWV={pwv}]')
```

# Raleigh-Jeans Temperature Brightness

```python
for nscale in NSCALES:
    data = np.load(f'{common_name}_{nscale}.out')
    plt.plot(data[:,0], data[:,2])

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'T$_b$ ' + f'[{T_unit}]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(NSCALES, title='NSCALE', loc=(1.01, 0.65))
plt.title(f'{title} [PWV={pwv}]')
```

# dTb / dZA

```python
for nscale in NSCALES:
    data = np.load(f'{common_name}_{nscale}.out')
    plt.plot(data[:,0], data[:,3])

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'$\partial$T$_b$ / $\partial$$\theta$ ' + f'[{T_unit} / {theta_unit}]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(NSCALES, title='NSCALE', loc=(1.01, 0.65))
plt.title(f'{title} [PWV={pwv}]')
```

# dTb / dPWV

```python
CORRECTION_FACTORS = np.array([436.075, 1083.1, 2109.86, 4367.34, 9713.75]) * 10**-3

for nscale, correction_factor in zip(NSCALES, CORRECTION_FACTORS):
    data = np.load(f'{common_name}_{nscale}.out')
    plt.plot(data[:,0], data[:,4] / correction_factor)

plt.xlim(min(data[:,0]), max(data[:,0]))
plt.ylabel(r'$\partial$T$_b$ / $\partial$PWV ' + f'[{T_unit} / {PWV_unit}]')
plt.xlabel(f'Frequency [{freq_unit}]')
plt.legend(NSCALES, title='NSCALE', loc=(1.01, 0.65))
plt.title(f'{title} [PWV={pwv}]')
```
