```python
%load_ext line_profiler
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 0.5, 100)
y = 0.25*np.random.randn(100) + np.linspace(0, 0.5, 100)

p = np.polyfit(x, y, 1)
plt.scatter(x, y)
# where ax + b
a, b = p
plt.plot(x, b + (a * x),'k--')
```

```python
p = np.polyfit(x, y, 3)
plt.scatter(x, y)
# where ax^3 + bx^2 + cx +d
a, b, c, d = p
plt.plot(x, d + (c * x),'k--')
```

```python
p = np.polyfit(x, y, 3)
plt.scatter(x, y)
# where ax^3 + bx^2 + cx +d
a, b, c, d = p
plt.plot(x, d*(x**0),'--', linewidth=3)
plt.plot(x, c*(x**1),'--', linewidth=3)
plt.plot(x, b*(x**2),'--', linewidth=3)
plt.plot(x, a*(x**3),'--', linewidth=3)
```
